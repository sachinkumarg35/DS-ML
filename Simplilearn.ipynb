# Gradient Descent

```python
cur_x = 5 # The algorithm starts at x=3
rate = 0.1 #learning rate
precision = 0.5 # thi tells us when to stop the algorithm
previous_step_size =1 #
max_iters = 1000000 #maximum number of iterations
iters = 0 #iteration counter

df = lambda x: 2+(x+5)# Gradient of our function

while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x #store current x in previous x
    cur_x = cur_x - rate * df(prev_x) # Grad descent
    previous_step_size = abs(cur_x - prev_x)#change in x
    print("iteration", iters, "\nx value is", cur_x) # print iterations
    
    print("the local minimum occurs at", cur_x)
```

# Statistics

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Name': ['Jane', 'Michael', 'willian', 'Rosy', 'Hana', 'ferdie', 'Gramo'],
                 'Salary':[50000, 54000, 50000, 189000, 55000, 40000, 59000]})
print(df)
```

```python
# Calulate the mean, median, mode
print('means', df['Salary'].mean())
print('median', df['Salary'].median())
print('mode', df['Salary'].mode())
print('min: '+str(df['Salary'].min()))
print('max: '+str(df['Salary'].max()))
print('range', df['Salary'].max()-df['Salary'].min())
```

```python
df['Salary'].describe()
```

```python
# Histogram of salaries
%matplotlib inline
import matplotlib.pyplot as plt

print(df)

salary = df['Salary']
salary.plot.hist(title='Salary Distribution', color='grey', bins=25)
plt.axvline(salary.mean(), color='violet', linestyle='dashed', linewidth=2)
plt.axvline(salary.median(), color='red', linestyle='dashed', linewidth=2)
plt.show()
```

# Probability


```python
# Creating a curly braces
Set1 = {4, 7}
print(Set1)
print(type(Set1))
```

```python
# Converting a list into set

my_list = [1,2,3,4,4]
my_set_from_list = set(my_list)
print(my_set_from_list)
```

```python
my_set = set([3,5,7])
print("Here is my set:", my_set)
print("3 is in the set:",3 in my_set)
print("1 is in the set:", 1 in my_set)
print("6 is not in the set:", 6 not in my_set)
```

```python
from itertools import product as prod

A = set([x for x in range(1, 7)])
B = set([x for x in range(1, 7)])
p=list(prod(A,B))

print("A is set of all possible outcomes of a die:", A)
print("B is set of all possible outcomes of a die:", B)
print("\nProduct of A andB is the all possible combinations of A and B thrown together:\n",p)
```

```python
#Throwing a 2 dice
n_dice = 2
dice_faces = (1,2,3,4,5,6)

#construct the event space i.e, set of all possible events
event_space = set(prod(dice_faces, repeat=n_dice))
event_space
```

```python
for outcome in event_space:
    print(outcome, end=', ')
print()
print(len(event_space))
```

```python
# Sum is a multiple of 3
favorable_outcome = []
for outcome in event_space:
    x,y = outcome
    if (x+y)%3==0:
        favorable_outcome.append(outcome)
favorable_outcome = set(favorable_outcome)

for f_outcome in favorable_outcome:
    print(f_outcome, end=', ')
print()
print(len(favorable_outcome))

prob = len(favorable_outcome)/len(event_space)
print("The probability of getting a sum which is a multiple of 3 is:", prob)
```

```python
# Throwing a 5 dice
n_dice = 5
dice_faces = {1,2,3,4,5,6}

#construct the event space i.e., set of all possible evets
event_space = set(prod(dice_faces, repeat=n_dice))
event_space
```

```python
len(event_space)
```

```python
# sum is a multiple of 5 but not a multiple of 3

favorable_outcome = []
for outcome in event_space:
    d1, d2, d3, d4, d5 = outcome
    if (d1+d2+d3+d4+d5)%5==0 and (d1+d2+d3+d4+d5)%3!=0:
        favorable_outcome.append(outcome)
favorable_outcome = set(favorable_outcome)
len(favorable_outcome)
```

```python
prob = len(favorable_outcome)/len(event_space)
print("The probability of getting a sum, which is a multiple of 5 but not a multiple of 3, is:", prob)
```
