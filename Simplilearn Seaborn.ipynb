```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Cars.csv')
df
```

```python
df.fillna(0.5)
```

```python
res = sns.barplot(df['animal'], df['age'])
plt.show()
```

```python
res = sns.barplot(df['animal'], df['age'], palette='Set2')
plt.show()
```

```python
res = sns.barplot(df['animal'], df['age'], palette='rocket')
plt.show()
```

```python
#countplot
sns.countplot(x = 'visits', data=df, palette='Set1')
plt.show()
```

```python
sns.countplot(x='animal', hue='age', data=df, palette='Set1')
plt.show()
```

```python
sns.distplot(df.age, bins=10, color='g')
```

```python
#Heatmap
sns.heatmap(df.corr(), cbar=True, linewidths=0.5)
```

```python
iris = pd.read_csv('iris.csv')
iris
```

```python
iris['Species'].value_counts()
```

```python
sns.scatterplot(x = 'Sepal.Length', y='Petal.Length', data=iris)
```

```python
sns.scatterplot(x='Sepal.Length', y='Petal.Length', data=iris, hue='Species', size='Species')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='Species',  palette="Set1")
```

```python
#Lmplot
sns.lmplot(x='Sepal.Length', y='Sepal.Width', data=iris)
```

```python
sns.lmplot(x='Sepal.Length', y='Sepal.Width', hue='Species', data=iris, markers=['o','*','^'])
```

```python
#Boxplot
sns.boxplot(x='Species', y='Sepal.Width', data=iris)
```
